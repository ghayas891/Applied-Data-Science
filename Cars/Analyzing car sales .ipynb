```python
selling_proportion = cars["brand"].value_counts(normalize = True)
selling_proportion
```

```python
selling_bol = selling_proportion > 0.05
selling_bol
```

```python
top_brand = selling_proportion[selling_bol]
```

```python
top_brand 
```

```python
top_selling_brand = top_brand.index
```

```python
top_selling_brand
```

```python
top_selling_brand_data = cars["brand"].isin(top_selling_brand)
```

```python
top_selling_brand_data
```

```python
top_selling_brand_cars = cars[top_selling_brand_data]
```

```python
top_selling_brand_cars
```

```python
top_selling_brand_cars = top_selling_brand_cars.copy()
```

```python
top_selling_brand_cars["price"] = top_selling_brand_cars["price"].str.replace("$","").str.replace(",","").astype(int)
```

```python
top_selling_brand_cars.head()
```

```python
top_selling_brand_cars.info()
```

```python
top_selling_brand_cars["odometer"] = top_selling_brand_cars["odometer"].str.replace("km","").str.replace(",","").astype(int)
```

```python
top_selling_brand_cars.rename({"price":"Price ($)","odometer":"odometer (KM)"},axis = 1,inplace = True)
```

```python
top_selling_brand_cars.head()
```

```python
grouping = top_selling_brand_cars.groupby("brand")
```

```python
grouping_mean = grouping["Price ($)","odometer (KM)"].mean()
grouping_mean
```

```python
grouping_mean.sort_values("Price ($)")
```
