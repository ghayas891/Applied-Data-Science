```python
def clean_date(date):
    bad_char = ["(",")","c","C","s",".","'"," "]
    for i in bad_char:
        date = date.replace(i,"")
    return date    
```

```python
date = "s.1967 C'"
clean_date(date)
```

```python
arts["Date"] = arts["Date"].apply(clean_date)
```

```python
arts["Date"].head(15)
```

```python
def split_date(date):
    if "-" in date:
        date = date.split("-")
        yr1 = int(date[0])
        yr2 = int(date[1])
        date = round((yr1+yr2)/2)
    else:
        date = int(date)
    return date    
```

```python
date = "1957-1948"
split_date(date)
```

```python
date = "2022"
split_date(date)
```

```python
arts["Date"] = arts["Date"].apply(split_date)
```
